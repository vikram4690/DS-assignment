```python
customers = pd.read_csv('drive/My Drive/vikii/data science project/Customers.csv')
```

```python
products = pd.read_csv('drive/My Drive/vikii/data science project/Products.csv')
```

```python
transactions = pd.read_csv('drive/My Drive/vikii/data science project/Transactions.csv')
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
merged_data = pd.merge(transactions, customers, on='CustomerID')
```

```python
merged_data.head()
```

```python
merged_data = pd.merge(merged_data, products, on='ProductID')
```

```python
merged_data.head()
```

```python
customer_product_matrix = merged_data.pivot_table(index='CustomerID', columns='ProductID', values='TotalValue', fill_value=0)
```

```python
customer_product_matrix.head()
```

```python
similarity_matrix = cosine_similarity(customer_product_matrix)
```

```python
similarity_df = pd.DataFrame(similarity_matrix, index=customer_product_matrix.index, columns=customer_product_matrix.index)
```

```python
similarity_df.head()
```

```python
lookalike_dict = {}
for customer_id in similarity_df.index:
    similar_customers = similarity_df.loc[customer_id].sort_values(ascending=False).iloc[1:4]
    lookalike_dict[customer_id] = list(similar_customers.items())
```

```python
lookalike_df = pd.DataFrame.from_dict(lookalike_dict, orient='index', columns=['Similar_Customer_1', 'Similar_Customer_2', 'Similar_Customer_3'])
```

```python
lookalike_df.head()
```

```python
lookalike_df.to_csv('Lookalike.csv', index=True)
```
